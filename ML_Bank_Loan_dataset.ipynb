#Decision Tree Classifier

```python
data_5_col.to_csv("Bank_loan_5_col.csv")
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
feature_col = ['Gender', 'Married', 'Self_Employed', 'Credit_History']
X = data_5_col[feature_col]
y = data_5_col['Loan_Status']
```

```python
X
```

```python
# Decision Tree Object
clf = DecisionTreeClassifier()

clf.fit(X,y)
```

```python
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,
                filled=True, rounded=True,
                special_characters=True, feature_names = feature_col,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('diabetes.png')
Image(graph.create_png())
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
X_train
```

```python
y_test.value_counts()
```

```python
y_train.value_counts()
```

```python
# Decision Tree Classifier Object
clf =  DecisionTreeClassifier()

clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
y_test.array
```

```python
y.array
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import (confusion_matrix, ConfusionMatrixDisplay, f1_score)
labels = [0,1]
cm = confusion_matrix(y_test, y_pred, labels=labels)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels = labels)
disp.plot();
```

```python
cm
```

```python
f1_score(y_test, y_pred)
```
